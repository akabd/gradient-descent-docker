```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Генерация случайных данных для "набора данных"
np.random.seed(42)
data_x = np.random.uniform(-2, 12, 100)
data_y = np.random.uniform(-2, 12, 100)

# Функция потерь на случайных данных
def loss_function_data(x, y, data_x, data_y):
    return np.mean((x - data_x)**2 + (y - data_y)**2)

# Градиенты функции потерь на случайных данных
def gradient_data(x, y, data_x, data_y):
    grad_x = 2 * np.mean(x - data_x)
    grad_y = 2 * np.mean(y - data_y)
    return grad_x, grad_y
```

```python
# Batch Gradient Descent
def batch_gradient_descent_real(start_x, start_y, learning_rate, iterations):
    x, y = start_x, start_y
    trajectory = [(x, y)]
    for _ in range(iterations):
        grad_x, grad_y = gradient_data(x, y, data_x, data_y)
        x = x - learning_rate * grad_x
        y = y - learning_rate * grad_y
        trajectory.append((x, y))
    return x, y, trajectory

# Начальные параметры
start_x, start_y = 10, 10
learning_rate = 0.1
iterations = 20

# Выполняем градиентный спуск
optimal_x, optimal_y, batch_trajectory = batch_gradient_descent_real(start_x, start_y, learning_rate, iterations)

# Печатаем результаты
print(f"Оптимальное значение x: {optimal_x}")
print(f"Оптимальное значение y: {optimal_y}")
print(f"Значение функции потерь в оптимальной точке: {loss_function_data(optimal_x, optimal_y, data_x, data_y)}")

# Визуализация траектории
def visualize_trajectory(data_x, data_y, batch_trajectory):
    plt.figure(figsize=(8, 6))
    plt.scatter(data_x, data_y, alpha=0.5, label='Random Data')  # Случайные данные
    trajectory_x, trajectory_y = zip(*batch_trajectory)          # Траектория оптимизации
    plt.plot(trajectory_x, trajectory_y, color='red', marker='o', label='Optimization Path')
    plt.title("Batch Gradient Descent Trajectory")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()

# Построение графика
visualize_trajectory(data_x, data_y, batch_trajectory)
```

```python
# Stochastic Gradient Descent
def stochastic_gradient_descent_real(start_x, start_y, learning_rate, iterations):
    x, y = start_x, start_y
    trajectory = [(x, y)]
    for _ in range(iterations):
        # Берём случайный пример из данных
        idx = random.randint(0, len(data_x) - 1)
        grad_x, grad_y = gradient_data(x, y, np.array([data_x[idx]]), np.array([data_y[idx]]))
        x = x - learning_rate * grad_x
        y = y - learning_rate * grad_y
        trajectory.append((x, y))
    return x, y, trajectory

# Начальные параметры
start_x, start_y = 10, 10
learning_rate = 0.1
iterations = 20

# Выполняем градиентный спуск
stochastic_x, stochastic_y, stochastic_trajectory = stochastic_gradient_descent_real(start_x, start_y, learning_rate, iterations)

# Печатаем результаты
print(f"Оптимальное значение x: {stochastic_x}")
print(f"Оптимальное значение y: {stochastic_y}")
print(f"Значение функции потерь в оптимальной точке: {loss_function_data(stochastic_x, stochastic_y, data_x, data_y)}")

# Извлекаем траекторию x и y
trajectory_x, trajectory_y = zip(*stochastic_trajectory)

# Визуализация
plt.figure(figsize=(8, 6))
plt.scatter(data_x, data_y, color='blue', alpha=0.5, label='Data Points')  # Данные
plt.plot(trajectory_x, trajectory_y, color='red', marker='o', label='Optimization Path')  # Траектория
plt.scatter(start_x, start_y, color='green', label='Start Point', zorder=5)  # Начальная точка
plt.scatter(stochastic_x, stochastic_y, color='purple', label='End Point', zorder=5)  # Конечная точка
plt.title('Stochastic Gradient Descent Trajectory')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
# Mini-Batch Gradient Descent
def mini_batch_gradient_descent_real(start_x, start_y, learning_rate, iterations, batch_size=10):
    x, y = start_x, start_y
    trajectory = [(x, y)]
    for _ in range(iterations):
        # Берём случайный мини-пакет данных
        indices = random.sample(range(len(data_x)), batch_size)
        grad_x, grad_y = gradient_data(x, y, data_x[indices], data_y[indices])
        x = x - learning_rate * grad_x
        y = y - learning_rate * grad_y
        trajectory.append((x, y))
    return x, y, trajectory

# Начальные параметры
start_x, start_y = 10, 10
learning_rate = 0.1
iterations = 20

# Выполняем градиентный спуск
mini_batch_x, mini_batch_y, mini_batch_trajectory = mini_batch_gradient_descent_real(start_x, start_y, learning_rate, iterations)

# Печатаем результаты
print(f"Оптимальное значение x: {mini_batch_x}")
print(f"Оптимальное значение y: {mini_batch_y}")
print(f"Значение функции потерь в оптимальной точке: {loss_function_data(mini_batch_x, mini_batch_y, data_x, data_y)}")

# Извлекаем траекторию x и y
trajectory_x, trajectory_y = zip(*mini_batch_trajectory)

# Визуализация
plt.figure(figsize=(8, 6))
plt.scatter(data_x, data_y, color='blue', alpha=0.5, label='Data Points')  # Данные
plt.plot(trajectory_x, trajectory_y, color='orange', marker='o', label='Optimization Path')  # Траектория
plt.scatter(start_x, start_y, color='green', label='Start Point', zorder=5)  # Начальная точка
plt.scatter(mini_batch_x, mini_batch_y, color='purple', label='End Point', zorder=5)  # Конечная точка
plt.title('Mini-Batch Gradient Descent Trajectory')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
